```python

#Importing the packages that will be used in this project


import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)



df = pd.read_csv(r'C:\Users\hehe\Desktop\movies.csv')
```

```python
df.head()
```

```python
#Looking for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Columns Data Types

df.dtypes
```

```python
df = df.dropna()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.title('Budget vs Gross Earnings')
plt.show()
```

```python
sns.regplot(x='budget',y='gross', data=df, line_kws={'color':'black'})
```

```python
#Correlation Matrix between all numeric columns

df.corr(method='pearson', numeric_only=True)
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
df.corr(method='kendall', numeric_only=True)
```

```python
correlation_matrix = df.corr(method='kendall', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
df.corr(method='spearman', numeric_only=True)
```

```python
correlation_matrix = df.corr(method='spearman', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
#Converting each categorical value into a unique numeric code using the `.cat.codes`


df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes 
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='spearman', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
corr_matrix = df_numerized.corr(method='pearson')
corr_pairs = corr_matrix.unstack()

corr_pairs
```
